```python
#Function to plot the data in the TSNE plotting manner
def tsne_plot(x1, y1):
    '''plot t-sne graph'''
    tsne = TSNE(n_components=2, random_state=0)
    X_t = tsne.fit_transform(x1)
    
    plt.figure(figsize=(12, 8))
    plt.scatter(X_t[np.where(y1==0), 0], X_t[np.where(y1==0),1], 
               marker='o',color='g', alpha=0.8, label='Non fraud')
    plt.scatter(X_t[np.where(y1==1), 0], X_t[np.where(y1==1),1], 
               marker='o',color='r', alpha=0.8, label='fraud')
    
    plt.legend(loc='best')
    plt.show()



rep_x = np.append(non_fraud,fraud, axis= 0)
y_n = np.zeros(non_fraud.shape[0])
y_f = np.ones(fraud.shape[0])
rep_y = np.append(y_n, y_f)


```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import confusion_matrix
```

```python
train_x,val_x,train_y,val_y=train_test_split(x,y,test_size=0.3)
```

```python
train_x=MinMaxScaler().fit_transform(train_x)
val_x=MinMaxScaler().fit_transform(val_x)
train_x=hidden_representation.predict(train_x)
val_x=hidden_representation.predict(val_x)
```

```python
parameters = {'solver':('lbfgs', 'liblinear'),'C':[0.2,0.5,1,10]}
lg=LogisticRegression()
tuned_lg=GridSearchCV(lg,parameters,cv=5)
tuned_lg.fit(train_x,train_y)
```

```python
clf = LogisticRegression(**tuned_lg.best_params_).fit(train_x, train_y)
pred_y = clf.predict(val_x)
```

```python
from sklearn.metrics import classification_report,accuracy_score
print ("")
print ("Classification Report: ")
print (classification_report(val_y, pred_y))

print ("")
print ("Accuracy Score: ", accuracy_score(val_y, pred_y))
```
